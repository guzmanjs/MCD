```python
# Otra forma de leer el archivo
# Lectura linea por linea

with open("D:\OneDrive - Tecnoquimicas\99. PERSONAL\Formación\Maestria\Semestre 1\Analisis Cuantitivo\Clase 3 - Practica\Medifis.csv", "r") as file:

    lines = list()

    for row in file.readlines():
        lines.append(row.replace('\n','').replace(',','.').split(';'))

col_names =['Genero','Avance','Estatura','Peso','LongitudPie','LongitudBrazo','AnchuraEspalda','DiametroCraneal','LongitudRodillaTobillo' ]

#Convertir en Dataframe a partir de la lista generada
df_medifis = pd.DataFrame(lines[8:], columns = col_names, dtype = float)
df_medifis.drop('Avance', axis=1, inplace=True)
df_medifis.head()
```
