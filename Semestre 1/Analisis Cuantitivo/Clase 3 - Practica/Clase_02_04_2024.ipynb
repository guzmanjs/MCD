## Modelo de Regresión Lineal

El modelo de regresión simple tiene la forma 
$$ y =b_0 + b_1*x + e $$

la recta de regresión 

$$ y = \^b_0+ \^b_1*x $$


```python
#Ejemplo 1

b_0 =1
b_1 = 3
n = 500 #Tamaño de muestra

x = np.random.normal(size=n) #Simulando n dayod notmslrd

e = np.random.normal(size= n, scale = 1.2)

y = b_0 + b_1*x + e

plt.scatter (x,y)
plt.title('Modelo de regresión simple')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
df = pd.DataFrame({'const': 1, 'x' : x, 'y' : y}) #Se crea la matrix incluyendo la columna de 1 que nos permitirá estimar el b0
df.head()
```

```python
X_design = df[['const', 'x']]
y = df[['y']]

left_term = np.linalg.inv(np.matmul(X_design.transpose(), X_design))
left_term

right_term = np.matmul(X_design.transpose(), y)
right_term

beta = np.matmul(left_term, right_term)

print(beta)
```

### Práctica cálculo de regresión lineal con matriz de covarianzas

$$ \hat{\beta_i} = \frac{Cov(X,Y)}{Var(X)}$$
$$ \beta_0 = \bar{Y} - \hat{\beta_i}*\bar{X}$$

```python
from sklearn import datasets
import pandas as pd
import numpy as np

iris = datasets.load_iris()

df = pd.DataFrame(iris.data, columns= iris.feature_names)
df.drop('sepal width (cm)', axis=1, inplace=True)
df.head()
```

```python
df.cov()
```

```python
Sigma = df.cov().to_numpy() #Matriz de covarianzas del dataset convertido a objeto Numpy

SigmaXX = Sigma[1:,1:] #Extraer la matriz de covarianzas de XX
SigmaXY =Sigma[1:,0] #Extraer la matriz de covarianzas de XY


#Estimacion de los betas de las variables de entrada
betas = np.matmul(np.linalg.inv(SigmaXX), SigmaXY) #np.linalg para Inversa de la matriz de covarianzas de XX
```

```python
betas
```

```python
#Para calcular el intercepto

y_bar = df['sepal length (cm)'].mean()

x_bar = df[['petal length (cm)', 'petal width (cm)']].mean()

beta0 = y_bar - np.matmul(x_bar.to_numpy(), betas)

beta0
```

Debemos validar los supuestos: 
1. Normalidad: Usamos Jarque - Bera, si el p_value es mayor a 0.05, no rechazo la hipótesis nula por tanto concluyo normalidad
$$ H_0 = X ~ Normal$$
$$ H_1 = X !~ Normal$$

2. Independencia: Usamos Durbin - Watson, si el valor DW está entre 1.5 y 2.5 decimos que hay independencia en los erroes, de lo contrario los errores no serian independientes.

3. Homocedasticidad: Usamos Breusch - Pagan test para el tema de homogeneidad de varianzas, si el p_value asociado al multiplicador de lagrange es mayor a 0.05 concluimos homocedasticidad, de lo contrario concluimos heterocedasticidad.
