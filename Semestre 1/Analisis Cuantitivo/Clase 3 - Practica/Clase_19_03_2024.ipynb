```python
### Otra forma de leer el archivo
## Esta celda lee todo el archi linea por linea
with open("Medifis.csv", "r") as file:
  lines = list()
  for row in file.readlines(): ### Este for recorre linea por linea el documento
    lines.append(row.replace("\n", "").replace(",", ".").split(";")) ### doy formato a cada linea y la agrego a la lista lines

col_names = ["genero", "Avance", "Estatura", "Peso", "LongitudPie", "LongitudBrazo",
             "AnchuraEspalda", "DiametroCraneal", "LongitudRodillaTobillo"]

df_medifis = pd.DataFrame(lines[8:], columns = col_names, dtype = float)
df_medifis.drop("Avance", axis = 1, inplace = True) ## .drop() me sirve para elimiar filas o columnas, en el caso de columnas axis = 1
df_medifis.head()
```

```python
import json ### Importamos el paquete json para manejo de archivos json

######## Las siguientes dos lineas me permiten leer el archivo con open()
with open("iris.json", "r") as archivo_iris:
  datos = archivo_iris.read() ### Lectura del archivo completo

datos_dic = json.loads(datos) ### transformando la cadena de strings en diccionarios
### Esto se hace para poder usarlo como un objeto tabular en pandas

df_iris = pd.DataFrame(datos_dic) ### contruimos el data frame con el diccionario
df_iris.head()
```

```python
### Esta celda la pueden usar para probar distintas contaminaciones
### y ver como se afecta la asimetria
from scipy.stats import skew
import matplotlib.pyplot as plt
clean = list(np.random.exponential(scale = 6, size = 500))

out = list(np.random.normal(loc = -800, size = 30))

x = clean + out
plt.hist(x)
skew(x)
```
