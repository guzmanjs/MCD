Si los datos que se van a usar para el modelo de regresiÃ³n se encuentran estandarizados o pasan por un preprocesamiento que los estandariza, podemos calcular los coeficientes de regresiÃ³n directamente con la matriz de correlaciones $C$ de la siguiente forma

$$\beta_{1\ldots p}=C_{XX}^{-1}C_Xy$$  
$$\beta_0 = 0 $$

```python
from sklearn.preprocessing import StandardScaler

scaler =StandardScaler() ##Esta linea instancia la clase que va a calcular la estandarizaciÃ³n

scaler.fit(df) ###Ajustamos el scaler a nuestros datos df

datos_scaled = scaler.transform(df) ##AsÃ­ calculo la estandarizaciÃ³n de los datos

df_scaled = pd.DataFrame(datos_scaled, columns= df.columns)

df_scaled.head()
```

```python
C = df_scaled.corr(method='pearson')

CXX = C.to_numpy()[1:, 1:]

CXy = C.to_numpy()[1:, 0]
```

```python
CXX
```

```python
import numpy as np

betas = np.matmul(np.linalg.inv(CXX), CXy)
betas
```

```python
import matplotlib.pyplot as plt

y_pred = np.matmul(df_scaled.to_numpy()[:, 1:], betas)

y_pred
```

```python
plt.scatter(y_pred, df_scaled.ventas.values)
```

```python
res = y_pred - df_scaled.ventas.values
plt.hist(res)
```
